```python

import numpy as np
import math

def angle_finder_shared(A, B, C, D):
    """
    Retrieves the angle between points, A,B,C,D by using the dot product. 

    When you have like a shared point, just use it again

    Args:
        A: (np.array) An array of size 3
        B: (np.array) An array of size 3
        C: (np.array) An array of size 3
        D: (np.array) An array of size 3
    
    Return:
        angle: (np.float) the angle between the 3 points
    """
    A_B = A-B
    D_C = C-D

    dotproduct = np.dot(A_B.T, D_C)
    A_Bmagnitude = np.linalg.norm(A_B)  
    D_Cmagnitude = np.linalg.norm(D_C)

    cos_theta = dotproduct /  (A_Bmagnitude * D_Cmagnitude)

    return np.arccos(cos_theta) * 180/math.pi

## Gonna calculate
# Neck angle: R shoulder, nose, L_shoulder
# Shoulder join angle: 
```

```python
## Body Velocity Variables:
# Need to groupby and then apply np.roll and then calculate it probably using next - old / 30 like the one above


def velocity_calc(data, column, groupby_col='trial_id'):
    """ 
    Does some quick velocity calculations for different bodyparts. 
    Follows this process:
    1. Groupby by the groupby object
    2. runs pd.shift() method for the column with the specific movement_axis (x, y, z, or all)
    3. differences the 2 
    4. divides by 30 to give us velocity
    
    Args:
        data (pd.DataFrame): the dataset
        column (list of strings): Columns
        groupby_col (string or list of strings): what to groupby by (eventually may have to factor multiple features)
    
    returns:
        Nonetype
    """
    group = data.groupby(groupby_col)
    
    for col in column:
        shifted_col = group[col].shift(1)
        data[f'VELOCITY_{col}'] = (data[col] - shifted_col) * 30
        
    return data
    


velocity_calc(data=data_copy, column = ['R_SHOULDER_x', 'R_SHOULDER_y', 'R_SHOULDER_z',
       'L_SHOULDER_x', 'L_SHOULDER_y', 'L_SHOULDER_z', 'R_ELBOW_x',
       'R_ELBOW_y', 'R_ELBOW_z', 'L_ELBOW_x', 'L_ELBOW_y', 'L_ELBOW_z',
       'R_WRIST_x', 'R_WRIST_y', 'R_WRIST_z', 'L_WRIST_x', 'L_WRIST_y',
       'L_WRIST_z', 'R_HIP_x', 'R_HIP_y', 'R_HIP_z', 'L_HIP_x', 'L_HIP_y',
       'L_HIP_z', 'R_KNEE_x', 'R_KNEE_y', 'R_KNEE_z', 'L_KNEE_x', 'L_KNEE_y',
       'L_KNEE_z', 'R_ANKLE_x', 'R_ANKLE_y', 'R_ANKLE_z', 'L_ANKLE_x',
       'L_ANKLE_y', 'L_ANKLE_z'])
```

```python
## Before we look at correlations of certain body parts based on the ball location. Let's start with simply calculating the XCF to understand 
# the xross correation function of the data. But I am not doing the entire unit root tests, yule walker stuff so might not be the most acc

## The idea is to calculate the cross correlation and then average it out across all 125 shots. We can do this with groupby!
# First adding a timestep column to make the data more timeseries

data_copy['FrameStep'] = data_copy.groupby(['trial_id']).cumcount() + 1  ## Not using this now but in the future
def calculate_XCF(group, col_1, col_2):
    group_normalized = (group[[col_1, col_2]] - group[[col_1, col_2]].mean()) / group[[col_1, col_2]].std()
    cross_corr = np.correlate(group_normalized[col_1], group_normalized[col_2], mode='full')
    
    return pd.Series(cross_corr)


groups = data_copy.groupby(['trial_id'])
cross_corr_results = data_copy.groupby(['trial_id']).apply(calculate_XCF, col_1='R_SHOULDER_x', col_2='R_ELBOW_x')
cross_corr_results_array = np.array(cross_corr_results)
mean_cross_corr = np.mean(cross_corr_results_array, axis=0)

lags = np.arange(-240 + 1, 240)
plt.plot(lags, mean_cross_corr)
plt.title('Average Cross-correlation (feature_1 vs feature_2) Across 125 Trials')
plt.xlabel('Lag')
plt.ylabel('Cross-correlation')
plt.show()
```
